## Checking and Handling Null Values

```python
data.isnull().sum()
```

Checking if all values in row are Null 

```python

data.isnull().all(axis=1)
```

```python
data.isnull().all(axis=1).sum()
```

Spouse index column has a lot of Null values. The column spouse index indicates if the customer is a spouse of an bank employee. Our target is to create 5 customer segments if we create a segment based on data if customer is a spouse of bank employee then we get a very small segment of total 1000000 customers. Hence, for our analysis now, we will drop the column of Spouse Index    

```python
# drop Spouse_index column
data.drop('Spouse_index', axis=1, inplace=True)
```

```python
data.head()
```

The other column that has more number of Null values is `Last_date_primary_cust`. 

The column indicates the last date of the customer being a Primary customer. On total 1102 customers had change in primary customer. Remaining 998899 customers were all primary customers. 

Also, the information if the customers had left their primary status after the first week can be obtained from `Primary_cust_index`. 
`Primary_cust_index` column indicates 1 if the primary customer and 99 is for primary customer after start of month. 

If the column `Last_date_primary_cust` is deleted then we lose the dates information of 1102 customers. Saving that information wont help ML model enough for training as well. Hence it is best to drop that column. 

```python
#drop Last_date_primary_cust column and print head
data.drop('Last_date_primary_cust', axis=1, inplace=True)
data.head()
```

```python
#print most frequent value of Province_name
data['Province_name'].value_counts().idxmax()
```

```python
# are the missing values of Province_name and Province__code_cust same?
data[data['Province_name'].isna()]['Province__code_cust'].isna().all()
```

```python
# fill missing values of Province_name with most frequent value
data['Province_name'] = data['Province_name'].fillna(data['Province_name'].value_counts().idxmax())
```

```python
# get province code for most frequent province name
data[data['Province_name'] == 'MADRID']['Province__code_cust'].value_counts().idxmax()
```

```python
# fill missing values of Province__code_cust with province code for most frequent province name
data['Province__code_cust'] = data['Province__code_cust'].fillna(data[data['Province_name'] == 'MADRID']['Province__code_cust'].value_counts().idxmax())
```
